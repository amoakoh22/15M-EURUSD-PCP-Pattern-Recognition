```python
# %% [code]
# Environment Setup: Install required libraries and mount Google Drive
!pip install mplfinance
!pip install yfinance
!pip install tensorflow
!pip install scikit-learn
!pip install seaborn
!pip install rouge_score
!pip install nltk
import nltk
nltk.download('punkt')

# %% [code]
# Import necessary libraries
import yfinance as yf
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix

# %% [code]
# Data Acquisition: Load XAUUSD M15 dataset from Google Drive
# Make sure your CSV file is located at '/content/drive/MyDrive/PCP Dataset/XAUUSD_M15.csv'
print("Loading dataset...")
ohlc_data = pd.read_csv('/content/drive/MyDrive/PCP Dataset/XAUUSD_M15.csv', sep='\t')
print("Initial Data:")
print(ohlc_data.head(10))

# %% [code]
# Data Cleaning & Preprocessing
# Clean column names (remove angle brackets)
ohlc_data.columns = [col.strip('<>') for col in ohlc_data.columns]

# Combine DATE and TIME columns into a single datetime column
ohlc_data['DATETIME'] = pd.to_datetime(ohlc_data['DATE'] + ' ' + ohlc_data['TIME'], format='%Y.%m.%d %H:%M:%S')

# Set the DATETIME column as the index and drop original DATE and TIME columns
ohlc_data.set_index('DATETIME', inplace=True)
ohlc_data.drop(columns=['DATE', 'TIME'], inplace=True)

# Drop any rows with missing data
ohlc_data.dropna(inplace=True)

print("\nCleaned Data:")
print(ohlc_data.head(10))
print("\nData Types:")
print(ohlc_data.dtypes)

# %% [code]
# Data Visualization: Plot OHLC data as a line plot and candlestick chart

# Reset index for mplfinance and rename columns accordingly
ohlc_data = ohlc_data.reset_index().rename(columns={'DATETIME': 'Datetime'})
ohlc_data['Datetime'] = pd.to_datetime(ohlc_data['Datetime'])
ohlc_data.set_index('Datetime', inplace=True)
ohlc_data = ohlc_data.rename(columns={'OPEN': 'Open', 'HIGH': 'High', 'LOW': 'Low', 'CLOSE': 'Close', 'VOLUME': 'Volume'})

# Plot line chart (using mplfinance "line" type)
mpf.plot(ohlc_data, type='line', style='charles', title='XAU/USD Line Chart', ylabel='Price', volume=False)

# Plot candlestick chart for full dataset
mpf.plot(ohlc_data, type='candle', style='charles', title='XAU/USD Candlestick Chart', ylabel='Price', volume=False)

# %% [code]
# Plot a candlestick chart for a specific subperiod (e.g., 29-31 December 2023)
start_date = '2023-12-29'
end_date = '2023-12-31'
sub_ohlc_data = ohlc_data[start_date:end_date].copy()
mpf.plot(sub_ohlc_data, type='candle', style='charles', title='XAU/USD Candlestick Chart (15Min) - 29 to 31 Dec 2023', ylabel='Price', volume=False)

# %% [code]
# Define the Deterministic PCP Logic to label the dataset.
def label_pcp(data):
    """
    Label data with:
      - 1 for Bullish PCP pattern,
      - 0 for Bearish PCP pattern,
      - -1 for No Pattern.
    """
    labels = np.full(len(data), -1)  # Default label: -1 (No Pattern)
    for i in range(2, len(data)):
        # Get Low and High values for three consecutive periods
        L2, L1, L0 = data['Low'].iloc[i-2:i+1]
        H2, H1, H0 = data['High'].iloc[i-2:i+1]
        # Bullish PCP condition
        if (L1 < L0) and (L1 < L2) and (H1 < H0) and (H1 < H2):
            labels[i] = 1  # Bullish
        # Bearish PCP condition
        elif (H1 > H0) and (H1 > H2) and (L1 > L0) and (L1 > L2):
            labels[i] = 0  # Bearish
    return labels

# Apply PCP labeling to the full dataset
ohlc_data['PCP_Label'] = label_pcp(ohlc_data.reset_index())

# Display a sample of labeled data
print("\nLabeled Data Sample:")
print(ohlc_data[['Open', 'High', 'Low', 'Close', 'PCP_Label']].head(20))

# %% [code]
# Visualize PCP signals on the candlestick chart
# Reset index to get Datetime as a column for plotting markers
ohlc_plot = ohlc_data.reset_index()

# Prepare marker Series for bullish and bearish signals
bullish_markers = pd.Series(np.nan, index=ohlc_plot.index)
bearish_markers = pd.Series(np.nan, index=ohlc_plot.index)
for i in range(len(ohlc_plot)):
    if ohlc_plot['PCP_Label'].iloc[i] == 1:  # Bullish PCP
        bullish_markers.iloc[i] = ohlc_plot['High'].iloc[i] + 0.0005
    elif ohlc_plot['PCP_Label'].iloc[i] == 0:  # Bearish PCP
        bearish_markers.iloc[i] = ohlc_plot['Low'].iloc[i] - 0.0005

# Create addplot markers for mplfinance
addplot = []
if not bullish_markers.isnull().all():
    addplot.append(mpf.make_addplot(bullish_markers, type='scatter', markersize=100, marker='^', color='green'))
if not bearish_markers.isnull().all():
    addplot.append(mpf.make_addplot(bearish_markers, type='scatter', markersize=100, marker='v', color='red'))

# Set 'Datetime' as index and plot the chart with PCP markers
ohlc_plot.set_index('Datetime', inplace=True)
mpf.plot(ohlc_plot, type='candle', style='charles', title='XAU/USD with PCP Markers', ylabel='Price', volume=False, addplot=addplot)

# %% [code]
# Filter out rows with "No Pattern" (-1) to retain only Bullish and Bearish signals
filtered_ohlc_data = ohlc_data[ohlc_data['PCP_Label'] != -1].copy()

# Show distribution of PCP labels
class_counts = filtered_ohlc_data['PCP_Label'].value_counts()
print("\nPCP Label Distribution:")
print(class_counts)

# Plot the distribution
plt.figure(figsize=(8, 5))
class_counts.plot(kind='bar', color=['red', 'green'])
plt.title('Distribution of PCP Labels (Bullish vs Bearish)')
plt.xlabel('Label (0: Bearish, 1: Bullish)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# %% [code]
# Prepare Data for Modeling
# Use only OHLC features to predict the PCP label (0 or 1)
X = filtered_ohlc_data[['Open', 'High', 'Low', 'Close']].values
y = filtered_ohlc_data['PCP_Label'].values

# Use TimeSeriesSplit for temporal cross-validation
tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# %% [code]
# Normalize the data using MinMaxScaler (you can also use StandardScaler)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Alternatively, using StandardScaler:
# std_scaler = StandardScaler()
# X_train = std_scaler.fit_transform(X_train)
# X_test = std_scaler.transform(X_test)

print("First five rows of normalized training data:")
print(X_train[:5])

# %% [code]
# Build the Neural Network Model using TensorFlow Keras
model = Sequential()
model.add(Input(shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(2, activation='softmax'))  # Output layer for 2 classes: Bearish and Bullish

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# %% [code]
# Train the model with EarlyStopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)
history = model.fit(X_train, y_train,
                    validation_split=0.2,
                    epochs=50,
                    batch_size=32,
                    callbacks=[early_stopping],
                    verbose=1)

# %% [code]
# Plot Training and Validation Accuracy and Loss
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs Epochs')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss vs Epochs')
plt.tight_layout()
plt.show()

# %% [code]
# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")

# %% [code]
# Generate Predictions on Test Data
y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)
print("Sample Predictions (first 10):")
print("Predicted:", y_pred[:10])
print("Actual:   ", y_test[:10])

# %% [code]
# Calculate and Display Classification Metrics
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Bearish', 'Bullish']))

conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=['Bearish', 'Bullish'], yticklabels=['Bearish', 'Bullish'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# %% [code]
# Save the Trained Model for future use
model.save('pcp_model_xauusd.keras')
print("Model saved successfully.")

# %% [code]
# (Optional) Load the saved model and generate predictions to simulate trade signals
loaded_model = load_model('pcp_model_xauusd.keras')
sample_prediction = loaded_model.predict(X_test[:5])
predicted_classes = np.argmax(sample_prediction, axis=1)
# Map predictions to trade signals: 1 -> "Buy", 0 -> "Sell"
trade_signals = ['Buy' if p == 1 else 'Sell' for p in predicted_classes]
print("Sample Trade Signals for first 5 test samples:")
print(trade_signals)
```
